```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
df = sns.load_dataset("iris")
```

```python
print("Dataset Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nSummary Statistics:\n", df.describe())
```

```python
sns.countplot(x="species", data=df, palette="Set2")
plt.title("Count of Each Iris Species")
plt.show()
```

```python
sns.boxplot(x="species", y="petal_length", data=df, palette="Set3")
plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
sns.violinplot(x="species", y="sepal_width", data=df, palette="Pastel1")
plt.title("Sepal Width by Species")
plt.show()
```

```python
sns.swarmplot(x="species", y="petal_width", data=df, palette="Dark2")
plt.title("Petal Width by Species")
plt.show()
```

```python
sns.kdeplot(x=df["sepal_length"], y=df["petal_length"], fill=True, cmap="Blues")
plt.title("Density Plot of Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
corr = df.drop(columns="species").corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Iris Features")
plt.show()
```
